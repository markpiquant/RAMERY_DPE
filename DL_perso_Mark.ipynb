```python
import matplotlib.pyplot as plt

# Calculer le nombre de valeurs non nulles par colonne
non_null_counts = df.count()

# Créer un histogramme
plt.figure(figsize=(10,6))
plt.bar(non_null_counts.index, non_null_counts.values)
plt.xlabel('Columns')
plt.ylabel('Non-null values count')
plt.title('Number of non-null values per column')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Calculer le seuil pour 80% de données non nulles
threshold = len(df) * 0.99

# Supprimer les colonnes avec moins de 80% de données non nulles
df_filtered = df.dropna(thresh=threshold, axis=1)

# Compter le nombre de colonnes restantes
num_columns_remaining = df_filtered.shape[1]

print(f"Number of columns remaining: {num_columns_remaining}")
```
